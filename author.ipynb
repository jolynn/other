# Extract author dates

```python
import requests
import pandas as pd
import numpy as np
import tqdm
from bs4 import BeautifulSoup
from collections import defaultdict

df = pd.read_csv('authorlist.csv')
df.head()
```

```python
def edit_name(name):
    if len(name.split()) > 1:
        return name.split()[0]+','+name.split()[1]
    else:
        return name

df['name'] = df.author.map(edit_name)
df.head()
```

```python
df.name
```

```python
s = requests.Session()
url = 'http://www.worldcat.org/identities/find?fullName='
new_data = defaultdict(list)


for i, row in tqdm.tqdm(df.iterrows()):
    full_url = url + row['name']
    data = s.get(url=full_url)
    response = BeautifulSoup(data.text)
    item = None
    
    if len(response.select('match[type="ExactMatches"]')) > 0:
    
        item = response.select('match[type="ExactMatches"]')[0]
         
            
    elif len(response.select('match[type="ExactMatches"]')) == 0:
        
        if len(response.select('match[type="ExactNameParts"]')) > 0:
            item = response.select('match[type="ExactNameParts"]')[0]
            #print('exact name parts:', row['name'])
        
        else:
            print('no match:', row['name'])        

        
     
    if item:
        new_data['title'].append(item.establishedform.text)
        
        if len(item.select('dates')) == 1:
            new_data['dates'].append(item.select('dates')[0].text)
        else:
            new_data['dates'].append(None)
    
        if len(item.select('pubdates')) == 1:
            new_data['pubdates'].append(item.select('pubdates')[0].text)
        else:
            new_data['pubdates'].append(None)
            
            
    else:
        new_data['title'].append(None)
        new_data['dates'].append(None)
        new_data['pubdates'].append(None)
```

```python
new_df = pd.concat([df.iloc[:633], pd.DataFrame(new_data)], axis=1)
new_df.shape
```

```python
new_df.head()
```

```python
def get_date(date, first=True):
    if pd.isnull(date):
        return None
    
    if len(date.split('-')) == 2:
        if first:
            return date.split('-')[0]
        
        else:
            return date.split('-')[1]
        
    elif len(date.split('-')) == 1:
        if first:
            return date
        else:
            return None
            

new_df['dates1'] = new_df.dates.map(get_date)
new_df['dates2'] = new_df.dates.map(lambda d: get_date(d, first=False))
new_df.head()
```

```python
new_df['pubdates1'] = new_df.pubdates.map(get_date)
new_df['pubdates2'] = new_df.pubdates.map(lambda d: get_date(d, first=False))
new_df.head()
```

```python
new_df.to_csv('authorlist_new.csv', index=False)
```
